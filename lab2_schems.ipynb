```python
import matplotlib.pyplot as plt
#plt.xkcd() #to make the diagram look hand drawn
%matplotlib inline
%config InlineBackend.figure_format = 'png'
import schemdraw 
import schemdraw.elements as elm
```

## Circuito inversor para una entrada


```python
with schemdraw.Drawing() as d:
    d += (op := elm.Opamp(leads=True).label('LM324',loc='bot'))
    d += elm.Line().down(d.unit/4).at(op.in2)
    d += elm.Ground(lead=False)
    d += (Rin := elm.Resistor().at(op.in1).left().idot().label('$1k\Omega$', loc='bot').label('$V_{2}$', loc='left'))
    d += (lin := elm.Line().up(d.unit/2).at(op.in1))
    d += (Rf := elm.Resistor().tox(op.out).label('$30k\Omega$'))
    d += elm.Line().toy(op.out).dot()
    d += elm.Line().right(d.unit/4).at(op.out).label('$V_{O}$', loc='right')
    d += (Rin2 := elm.Resistor().at(Rf.start).left().idot().label('$1k\Omega$', loc='bot').label('$V_{1}$', loc='left'))
```

```python
with schemdraw.Drawing() as d:
    d += (op := elm.Opamp(leads=True).label('LM324',loc='bot'))
    d += elm.Line().right(d.unit/4).at(op.out).label('$V_{O}$', loc='right')
    d += elm.Line().down(d.unit/4).at(op.in2)
    d += elm.Ground(lead=False)
    d += (Rin := elm.Resistor().at(op.in1).left().idot().label('$R$', loc='bot').label('$V_{i}$', loc='left'))
    d += (Rf := elm.Resistor().up().at(op.in1).label('$R_f$'))
    d += elm.Line().down(d.unit/4).at(op.in1)
    d += elm.Ground(lead=False)
    d += (V := elm.SourceV().right().at(Rf.end).reverse().label('$V_{O}$', loc='top'))
    d += elm.Line().down(d.unit/4)
    d += elm.Ground(lead=False)
```
